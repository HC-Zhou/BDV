```python
import matplotlib.pyplot as plt
import numpy as np
```

- 使用linspace函数定义x为0.5到3.5之间产生的100个数据，y = sin(x)，绘制曲线，如图蓝色线条；

```python
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
plt.plot(x, y, 'b')
```

- 使用random函数定义y1表示随机产生的100个数，绘制散点，如图蓝色星花;

```python
y1 = 6 * np.random.random(100) - 3
plt.scatter(x, y1, marker='p', color='b')
```

- 使用xlim和ylim函数，显示x轴区间为（0，4），y轴区间为（-3，3);
- 显示网格线;
- 显示x、y轴标签;
- 绘制水平参考线，如图红色线条;
- 绘制竖直参考区间，如图黄绿色区域;
- 绘制水平参考区域，如图绿色区域;
- 绘制指向型文本注释，如图蓝色字体，maximum;
- 绘制无指向型文本注释，如图蓝色字体，y=sin(x);
- 添加title及文本标签图例;

```python
plt.figure(figsize=(8, 6), dpi=85)
plt.plot(x, y, 'b', label='plot figure')
plt.scatter(x, y1, marker='p', color='b', label='scatter figure')
plt.xlim([0, 4])  # x轴区间
plt.ylim([-3, 3])  # y轴区间
plt.xlabel('x-axis')  # x轴标签
plt.ylabel('y-axis')  # y轴标签
plt.grid(color='pink', linestyle='--')  # 网络线
plt.axhline(y=0.5, xmin=0.5, color="red", linestyle="--")  # 画出参考线
plt.axvspan(xmin=1.0, xmax=3.0, facecolor="y", alpha=0.3)  # 竖参考区间
plt.axhspan(ymin=0.0, ymax=1.0, facecolor="g", alpha=0.2)  # 横参考区间
plt.annotate('maximum', color="dodgerblue", xy=(1.7, 1), xytext=(2.7, 0.7),
             arrowprops=dict(color='dodgerblue', headwidth=8, width=0.8))  # 绘制指向型文本注释，如图蓝色字体，maximum
plt.annotate("y = sin(x)", xy=(3.3, 0), xytext=(0, 0), textcoords='offset points',
             color='b')  # 绘制无指向型文本注释，如图蓝色字体，y=sin(x);
plt.title("y = sin(x)")  # 添加title及文本标签图例
plt.legend(loc='upper right')
plt.savefig('项目1.png')
plt.show()
```
